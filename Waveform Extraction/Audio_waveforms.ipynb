```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

def plot_trimmed_waveform(audio_file_path, duration=30):
    audio, sr = librosa.load(audio_file_path)

    samples_to_keep = int(duration * sr)

    trimmed_audio = audio[:samples_to_keep]

    plt.figure(figsize=(12, 4))
    librosa.display.waveshow(trimmed_audio, sr=sr)
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.title("Audio Waveform of Tap")
    plt.show()

audio_file_path = '/content/drive/MyDrive/Summer_Internship_Data/Processed_audios/Tap1_processed.wav'

plot_trimmed_waveform(audio_file_path, duration=30)
```
